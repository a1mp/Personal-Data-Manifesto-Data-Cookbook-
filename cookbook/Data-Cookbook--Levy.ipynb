# Fourth,my favourite part, Data wrangling and Visualizations

My interest is to analyze a country's data through visualizations.
For that I just to choose to write a function that can be given a country and do the Job 

```python
# Before cutting the cheese, let's just plot a scatter of everything

#assign the figure size and the axes point for the scatter plot
fig, axs = plt.subplots(figsize=(12, 8))

df['Literacy rates (World Bank, CIA World Factbook, and other sources)'].plot(kind='line', rot=60, ax=axs)
#Lables and titles
plt.xlabel("X_axis")
plt.ylabel("Literacy Rate")
plt.title('World Literacy rates')

#to print your graph
plt.show()
```

We can't say much when plotted like this, it is better therefore to do some wrangling before rushing to scattering the data

```python
# I like to sum up everything in a simple function
def literacy_rates_plot(country):
    # Opening the data frame
    df = pd.read_csv('cross-country-literacy-rates.csv')
    
    # Check if the country is in the data frame
    if country not in df['Entity'].unique():
        print(f"{country} is not present in the data frame.")
        return

    # Isolate dataframe for one country using .loc[] 
    df_country = df.loc[df['Entity'] == country]

    # Select only the columns you want to keep.
    df_reduced_country = df_country.loc[:, ['Year', 'Literacy rates (World Bank, CIA World Factbook, and other sources)']]

    # Plot the data set
    plt.plot(df_reduced_country['Year'], df_reduced_country['Literacy rates (World Bank, CIA World Factbook, and other sources)'], color='red', label='Data from our world data')

    # Add labels and title to the plot
    plt.xlabel('Year')
    plt.ylabel('Literacy rates (World Bank, CIA World Factbook, and other sources)')
    plt.title(f'{country} - Data Set')

    # Add a legend to the plot
    plt.legend()

    # Show the plot
    plt.show()
```

we could use powerful tools such as flourish https://flourish.studio/ and observable https://observablehq.com/to see what it can generate. I used flourish on this dataframe to produce a visualization that appears in my personal manifesto.

```python
literacy_rates_plot('Kazakhstan')
```

```python
literacy_rates_plot('Burundi')
```

```python
literacy_rates_plot('World')
```

For the few examples, we can see that there was an increase in literacy rates for Burundi (my home), Khazakhstan, and an overall increase for the world. Aren't you curious to try that of your country?

```python
# Let's type a country that doesn't exist
literacy_rates_plot('Agahugu')
```

I scrapped the Whitman athletic website with help

```python
#Another useful library
import requests
import json
from bs4 import BeautifulSoup
```

I scrapped the Whitman athletic website

```python
whitman_url= 'https://athletics.whitman.edu/sports/swimming-and-diving'
w_resp= requests.get(whitman_url)
soup=BeautifulSoup(w_resp.content, 'html.parser')
soup
```

```python
# Find all div elements with class "row"
link= soup.find_all('div',{"class":"row"})
# Print the text of each div element
for l in link:
    print(l.text)
```

```python
# Find the "Roster" link by its text content
roster_link = soup.find('a', text='Roster')

# Extract the URL of the "Roster" link
if roster_link:
    roster_url = roster_link.get('href')
    print(roster_url)
else:
    print('Roster link not found.')
```
